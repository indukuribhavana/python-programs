```python
#TUPLES
a={"divija","abhi","abhiram","akhila"}
print(a)

#ACCESS TUPLE ITEMS
names=("divija","abhi","abhiram","akhila")
print(names[1])

##NEGATIVE INDEXING
names=("divija","abhi","abhiram","akhila")
print(names[-1])

##RANGE OF INDEXES
names=("divija","abhi","abhiram","akhila")
print(names[1:3])

names=("divija","abhi","abhiram","akhila")
print(names[:2])

names=("divija","abhi","abhiram","akhila")
print(names[2:])

names=("divija","abhi","abhiram","akhila")
print(names[-1:-3])

names=("divija","abhi","abhiram","akhila")
if "divija" in names:
  print("yes,'divija' is in the name tuple")


#UPDATE TUPLES
##CHANGE TUPLE VALUES
names=("divija","abhi","abhiram")
b=list(names)
b[2]="akhila"
names=tuple(b)
print(names)

##ADD ITEMS
a=("divija","abhi","abhiram")
b=list(a)
b.append("akhila")
a=tuple(b)

a = ("divija","abhi","abhiram")
b = ("akhila",)
a +=b
print(a)


##REMOVE ITEMS
a=("divija","abhi","abhiram","akhila")
b=list(a)
b.remove("abhi")
a=tuple(b)
print(a)


#UNPACK TUPLES
names = {"divija","abhi","abhiram","akhila"}
(white,yellow,blue,black) = names
print(white)
print(yellow)
print(blue)

##USING ASTERISK
names = {"divija","abhi","abhiram","akhila"}
(white,yellow,blue,*black) = names
print(white)
print(yellow)
print(blue)


#LOOP TUPLES
names = {"divija","abhi","abhiram","akhila"}
for a in names:
  print(a)

##LOOP THROUGH INDEX NUMBERS
names = ("divija","abhi","abhiram","akhila")  
for i in range(len(names)):
  print(names[i])


##USING WHILE LOOP
names = ("divija","abhi","abhiram","akhila")
i = 0
while i < len(names):
  print(names[i])
  i = i + 1


#JOIN TUPLES
tuple1 = ("divija","abhi","abhiram","akhila")
tuple2 = (45,67,89)
tuple3 = tuple1 + tuple2
print(tuple3)

##MULTIPLY TUPLES
names = ("divija","abhi","abhiram","akhila")
mytuple = names * 2
print(mytuple)
```
