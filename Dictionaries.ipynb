```python
##PYTHON DICTIONARIES
choclate= {
    "brand": "cadbury",
    "model": "oreo",
    "year":2022
}


choclate= {
    "brand": "cadbury",
    "model": "oreo",
    "year":2022
}
print(choclate)

##DICTIONARY ITEMS
choclate= {
    "brand": "cadbury",
    "model": "oreo",
    "year": 2022
}
print(choclate["brand"])

choclate= {
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
    "year":2020
}
print(len(choclate))

choclate= {
    "brand": "cadbury",
    "electric": "false",
    "year":2022,
    "colors":["black","white","red"]
}

##TYPE
choclate= {
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
    "year":2020
}
print(type(choclate))    


#ACCESS DICTIONARY
myorder={
    "model":"oreo",
    "brand":"cadbury",
    "item":"choclate"
}
a=myorder.get("model") 
print(a)   

#get keys
myorder={
    "model":"oreo",
    "brand":"cadbury",
    "item":"choclate"
}
a=myorder.keys() 
print(a)   

choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
}
a=choclate.keys()
print(a)
choclate["color"]="brown"
print(a)

#get values
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
}
a=choclate.values()
print(a)
choclate["color"]="brown"
print(a)

#get items
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
    "color":"brown"
}
a=choclate.items()
print(a)

#check if key exists
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
}
if "model" in choclate:
  print("yes,'model' is one of the keys in the choclate dictionary")    


#CHANGE VALUES
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
    "color":"brown"
}
choclate["year"]=2022
print(choclate)

choclate.update({"color":"brown"})
print(choclate)


#ADDING ITEMS
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022
}
choclate["color"] ="brown"
print(choclate)


#REMOVING ITEMS
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":2022,
    "color":"brown"
}
print(choclate)
choclate.pop("color")
print(choclate)
choclate.popitem()
print(choclate)
del choclate["brand"]
print(choclate)


#LOOP DICTIONARIES
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":"2022",
    "color":"brown"
}
for a in choclate:
  print(a)

for a in choclate:
  print(choclate[a])

for a in choclate.values():
  print(a)

for a in choclate.keys():
  print(a)

for a,b in choclate.items():
  print(a,b)


#COPY DICTIONARIES
choclate={
    "brand": "cadbury",
    "model": "oreo",
    "year":"2022",
    "color":"brown"
}
print(choclate)
a=choclate.copy()
print(a)

b=dict(choclate)
print(b)


#NESTED DICTIONARIES
myfriends={
      "friend1":{
      "name":"divija",
      "year":2003
  },
   "friend2":{
      "name":"abhi",
      "year":2003
  },
  "friend3":{
      "name":"abhiram",
      "year":2004
  },
  "friend4":{
      "name":"akhila",
      "year":2004
  }
}
print(myfriends)
```
